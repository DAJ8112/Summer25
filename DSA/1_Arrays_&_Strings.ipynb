### **Arrays**

```python
A = [1, 2, 3]

print(A)
```

```python
# Append - Insert element at end of array - On Average: O(1)
A.append(4)

print(A)
```

```python
# Pop - Deleting element at end of array - O(1)
A.pop()

print(A)
```

```python
# Insert (not at end of Array) - O(n)
A.insert(2, 8)

print(A)
```

```python
# Modify an element - O(1)

A[0] = 7

print(A)
```

```python
# Accessing an element at given index - O(1)

print(A[2])
```

```python
# Checking if array has an element - O(n)

if 6 in A:
    print(True)
else:
    print(False)
```

```python
# Check the length - O(1)

print(len(A))
```

### **Strings**

```python
# Append to end of string - O(n)

s = 'hello'

b = s + 'z'

print(b)
```

```python
# Check if something is in string - O(n)
if 'l' in s:
    print(True)
```

```python
# Accessing specific positions - O(1)

print(s[3])
```

```python
# Check the length - O(1)

print(len(s))
```
